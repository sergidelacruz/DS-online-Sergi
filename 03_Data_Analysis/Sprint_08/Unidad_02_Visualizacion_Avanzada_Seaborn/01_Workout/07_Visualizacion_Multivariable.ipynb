### Mapas de Calor

Recuperamos rapidamente lo que ya vimos en la unidad de anlisis multivariante mostrando el mapa de calor de la matriz de correlacin de variables dos a dos.

```python
matriz_corr = df_vuelos.corr(numeric_only= True)
```

```python

plt.figure(figsize=(5,5))
sns.heatmap(matriz_corr,
            vmin=-1,
            vmax=1,
            cmap=sns.diverging_palette(145, 280, s=85, l=25, n=7),
            square=True,
            linewidths=.1,
            annot=True);
```

Recuerda que si ya es difcil ver mapas de calor con las tablas de contingencia de variables categricas de ms de 3 valores por variable, cuando tengamos ms de 3 variables numricas a comparar se va a hacer igual de difcil. Despeja el mapa:

```python

plt.figure(figsize=(5,5))
sns.heatmap(matriz_corr,
            vmin=-1,
            vmax=1,
            cmap=sns.diverging_palette(145, 280, s=85, l=25, n=7),
            square=True,
            linewidths=.1,
            annot=False);

plt.axis("off")
```

El mensaje: el dataset tiene las variables altamente correladas entre s (casi ningun valor debajo del 0)


### Burbujas con 3 y 4 dimensiones


Para casi terminar, los diagramas de burbujas permiten mostrar hasta cuatro variables: una en x, otra en y, otra asociada al tamao de los puntos y otra asociada a los colores

#### Burbujas con 3 dimensiones: Ingresos, distancia, consumo:

```python
sns.scatterplot(x="distancia", y= "consumo_kg", size = df_vuelos["ingresos"]/1000, data = df_vuelos);
```

#### Burbujas con 4 dimensiones: Ingresos, distancia, consumo por compaa area

```python
sns.scatterplot(x="distancia", y= "consumo_kg", size = df_vuelos["ingresos"]/1000, data = df_vuelos, hue = "aircompany");
```

Para el anlisis pueden servirnos bastante como discriminador y gua para ahondar en anlisis ms dirigidos. Como visaulizacin descriptiva resumen mucha informacin, pero... para presentaciones... tienes que manejarlos con cuidado (o hacer la presentacin solo en torno a este diagrama)
