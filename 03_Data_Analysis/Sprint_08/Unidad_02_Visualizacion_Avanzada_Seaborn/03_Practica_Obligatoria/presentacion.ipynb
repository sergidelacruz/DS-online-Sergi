1. El titanic fue un accidente con una tasa de mortalidad alta.  
2. Los datos apoyan que se dio el "Las mujeres y los niños primero", pero...  
3. ...la clase en la que viajases tuvo una influencia significativa y podría explicar...  
4. ...las diferencias entre tasas de supervivencia de mujeres y niños.    
5. Además, aunque el puerto de embarque parece tener influencia, se podría asumir que más bien fue la clase de pasaje que embarcó en cada uno de ellos. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import bootcampviztools
```

```python
df = pd.read_csv("./data/titanic.csv")
```

```python
valores = df.alive.value_counts(normalize=True) * 100
sns.barplot(x=valores.index, y=valores.values, hue = valores.index)
plt.title("Tasa de mortalidad del Titanic")
plt.ylabel("%")
plt.savefig("./img/imagen_1.png")
```

```python
valores = df.who.value_counts(normalize=True) * 100
sns.barplot(x=valores.index, y=valores.values, hue = valores.index)
plt.title("Tasa de mortalidad según género")
plt.ylabel("%");
plt.savefig("./img/imagen_2.png")
```

```python
valores = df["class"].value_counts(normalize=True) * 100
sns.barplot(x=valores.index, y=valores.values, hue = valores.index)
plt.title("Tasa de mortalidad según clase")
plt.ylabel("%");
plt.savefig("./img/imagen_3.png")
```

```python
df_women = df[(df["who"] == "woman") & (df["alive"] == "no")]
```

```python
sns.catplot(x="alive",
            col = "who",
            kind="count",
            edgecolor=".6",
            hue = "class",
            legend= True,
            data=df_women);
plt.title("Mortalidad mujeres según clase");
plt.savefig("./img/imagen_4.png")
```

```python
df_cher = df[(df["embark_town"] == "Cherbourg")]
```

```python
sns.catplot(x="alive",
            col = "embark_town",
            kind="count",
            edgecolor=".6",
            hue = "alive",
            legend= True,
            data=df_cher,
            );
plt.title("Tasa de mortalidad de los pasajeros de Cherbourg")
plt.savefig("./img/imagen_5_1.png")
```

```python
sns.catplot(x="class",
            col = "embark_town",
            kind="count",
            edgecolor=".6",
            orient = "V",
            hue = "class",
            legend= True,
            data=df_cher);
plt.title("Distribución según Clase de los pasajeros de Cherbourg")
plt.savefig("./img/imagen_5_2.png")
```
