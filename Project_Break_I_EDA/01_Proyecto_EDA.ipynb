## 1. Planteamiento del problema e hiptesis

El Anlisis Exploratorio de Datos (EDA) es el proceso de examinar, visualizar y resumir un conjunto de datos para descubrir patrones, detectar anomalas y validar suposiciones. Es un paso fundamental antes de aplicar tcnicas de Machine Learning o tomar decisiones basadas en datos.

### Vuestro objetivo

Resolver un problema real mediante el anlisis de datos siguiendo este proceso:

1. **Definir el problema** - Qu queris descubrir o demostrar?
2. **Plantear hiptesis** - Formulad suposiciones verificables antes de explorar los datos
3. **Analizar los datos** - Explorad, visualizad y probad vuestras hiptesis
4. **Extraer conclusiones** - Verificad si vuestras hiptesis se cumplieron o descubristeis algo diferente

### Formulacin de hiptesis

Plantead **una o dos hiptesis principales** que guen vuestro anlisis. Una hiptesis debe ser:

- **Especfica:** Clara y concreta, no ambigua
- **Medible:** Cuantificable con los datos disponibles
- **Verificable:** Posible de confirmar o refutar mediante anlisis

**Ejemplos de hiptesis bien formuladas:**

- "Los pasajeros de primera clase tienen una tasa de supervivencia significativamente mayor que los de segunda y tercera clase"
- "Existe una diferencia significativa en el precio por metro cuadrado entre viviendas ubicadas en zonas cntricas y perifricas"
- "Existe una relacin positiva entre el ndice de Masa Corporal (IMC) y la probabilidad de diagnstico de diabetes"

### Consejos importantes

> **Consejo #1:** Un buen EDA sienta las bases para futuros proyectos de Machine Learning. Si vuestros datos tienen una variable objetivo clara (estn etiquetados), el anlisis ser ms enfocado y los resultados ms aplicables a modelos predictivos.

> **Consejo #2:** Evitad datasets que requieran conocimiento tcnico muy especializado o anlisis estadsticos avanzados que estn fuera del alcance de un EDA bsico.
>
> **Ejemplos de temas que suelen ser problemticos:**
> - **Mercados financieros:** Requieren anlisis de series temporales complejas, conocimientos de trading y modelos economtricos especficos
> - **Deportes de alto rendimiento:** Los datos suelen estar agregados de formas que ocultan patrones individuales y requieren mtricas especializadas del deporte
> - **Seales mdicas continuas (ECG, EEG):** Necesitan procesamiento de seales y conocimiento clnico especializado
>
> **Nuestra recomendacin:** Elegid datasets con observaciones individuales claras, variables comprensibles y problemas que podis interpretar sin ser expertos/as en el dominio.

> **Consejo #3:** No est permitido usar datasets "famosos" o de prctica comn como Titanic, Iris, Boston Housing o California Housing. Buscad un dataset original que os motive y permita realizar un anlisis interesante y diferenciador.
