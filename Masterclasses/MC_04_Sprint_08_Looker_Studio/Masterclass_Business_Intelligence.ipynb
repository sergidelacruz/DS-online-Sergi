## ¿Qué es Business Intelligence (BI)?

> **Business Intelligence (BI)** es un conjunto de procesos tecnológicos para recopilar, gestionar y analizar datos organizacionales con el fin de generar información que sirva de base para estrategias y operaciones comerciales.

Fuente: [IBM](https://www.ibm.com/topics/business-intelligence)

### En términos prácticos, BI responde preguntas como:
- ¿Cuáles fueron las ventas del último trimestre por región?
- ¿Qué productos tienen mejor margen de rentabilidad?
- ¿Cómo evoluciona la retención de clientes mes a mes?
- ¿Dónde están los cuellos de botella en nuestros procesos?

## Sistemas de Información y BI

Un **Sistema de Información (SI)** es un conjunto de elementos organizados para tratar datos e información con un propósito específico.

### Los SI tienen dos objetivos principales:

| Objetivo | Descripción | Ejemplo |
|----------|-------------|----------|
| **Operacional** | Soporte informático para procesos básicos de gestión | ERP, CRM, sistemas de facturación |
| **Analítico** | Soporte para toma de decisiones estratégicas | Dashboards, reportes, predicciones |

**Business Intelligence se enfoca en el segundo objetivo**: convertir datos operacionales en información útil para la toma de decisiones.

## Business Intelligence como visión estratégica

BI no es solo una herramienta, es una **visión estratégica** de cómo una organización gestiona y aprovecha sus datos.

### Componentes principales de BI:

| Componente | Función |
|------------|----------|
| **Data Warehouses** | Almacenes centralizados de datos históricos |
| **Herramientas OLAP** | Análisis multidimensional de datos |
| **Reporting y Dashboards** | Visualización y monitorización |
| **Minería de datos** | Descubrimiento de patrones ocultos |

## Arquitectura del BI

Un sistema de BI típico se organiza en **tres niveles**:

| Nivel | Componentes | Función |
|--------|-------------|----------|
| **Top Tier** | Reporting, Dashboards, Analytics, Minería de datos | Donde los usuarios interactúan con los datos. *Looker Studio opera aquí.* |
| **Middle Tier** | Servidores OLAP | Procesa consultas analíticas y permite análisis multidimensional rápido |
| **Bottom Tier** | Bases de datos, ETL, Data Warehouse, Data Marts | Almacena y prepara los datos desde fuentes operacionales |

## Herramientas y operadores OLAP

**OLAP (Online Analytical Processing)** permite el análisis multidimensional de los datos, como si fueran un "cubo" que podemos rotar y explorar.

### Operadores principales:

| Operador | Función | Ejemplo práctico |
|----------|---------|------------------|
| **Drill-down** | Aumentar el nivel de detalle | De ventas anuales → ventas mensuales → ventas diarias |
| **Roll-up** | Agrupar o resumir datos | De ventas por tienda → ventas por región → ventas totales |
| **Slice** | Seleccionar una "rebanada" del cubo | Ver solo datos del año 2024 |
| **Dice** | Seleccionar un subcubo | Ventas de electrónica en Madrid durante Q1 |
| **Pivot** | Rotar la perspectiva de análisis | Cambiar filas por columnas en una tabla |

## KPIs (Key Performance Indicators)

Antes de hablar de dashboards, es fundamental entender qué son los **KPIs**.

> Un **KPI** es una métrica cuantificable que refleja el rendimiento de una organización en relación con sus objetivos clave.

### Características de un buen KPI:
- **Específico**: Mide algo concreto
- **Medible**: Se puede cuantificar
- **Alcanzable**: Es realista
- **Relevante**: Está alineado con objetivos de negocio
- **Temporal**: Tiene un marco de tiempo definido

### Ejemplos de KPIs por área:

| Área | KPI | Fórmula |
|------|-----|----------|
| Ventas | Tasa de conversión | (Ventas / Visitas) × 100 |
| Marketing | CAC (Coste de Adquisición) | Gasto marketing / Nuevos clientes |
| Finanzas | Margen bruto | (Ingresos - Costes) / Ingresos × 100 |
| RRHH | Rotación de personal | Bajas / Plantilla media × 100 |

## Dashboards vs Informes

Aunque a veces se usan indistintamente, dashboards e informes tienen propósitos diferentes:

| Característica | Dashboard | Informe |
|----------------|-----------|----------|
| **Enfoque** | KPIs y visión global | Datos detallados y análisis profundo |
| **Formato** | Visual, conciso, interactivo | Extenso, puede incluir dashboards |
| **Objetivo** | Supervisar de un vistazo | Documentar y explicar información |
| **Actualización** | Tiempo real o frecuente | Periódica (semanal, mensual) |
| **Audiencia** | Ejecutivos, gestores | Analistas, equipos técnicos |
| **Interactividad** | Alta (filtros, drill-down) | Baja o media |

### ¿Cuándo usar cada uno?

**Usa un Dashboard cuando necesites:**
- Monitorizar KPIs en tiempo real
- Tomar decisiones rápidas
- Comunicar estado general a stakeholders

**Usa un Informe cuando necesites:**
- Documentar análisis detallados
- Explicar el "por qué" detrás de los números
- Presentar conclusiones y recomendaciones

## Comparación entre herramientas de BI

| Herramienta | Facilidad de uso | Integración nativa | Visualizaciones | Licencia | Ideal para |
|-------------|------------------|-------------------|-----------------|----------|------------|
| **Power BI** | Alta | Microsoft 365 | Buena | Pago (versión gratuita limitada) | Usuarios de ecosistema Microsoft |
| **Tableau** | Media | Múltiples conectores | Excelente | Pago | Análisis visual avanzado |
| **Qlik Sense** | Media | Moderada | Buena | Pago | Análisis asociativo |
| **Looker Studio** | Alta | Google Workspace | Moderada | **Gratuita** | Usuarios de Google, presupuesto limitado |

### ¿Por qué elegir Looker Studio?

**Gratuito**: Sin costes de licencia  
**Integración Google**: Conexión nativa con Google Analytics, Sheets, BigQuery  
**Colaborativo**: Compartir y editar como Google Docs  
**Fácil de aprender**: Interfaz intuitiva de arrastrar y soltar  
**En la nube**: Sin instalación, acceso desde cualquier lugar  

**Limitaciones a considerar:**
- Visualizaciones menos avanzadas que Tableau
- Mejor rendimiento con datos en ecosistema Google
- Funcionalidades de modelado de datos básicas

## Ciclo de vida de un proyecto BI

Un proyecto de Business Intelligence típicamente sigue estas fases:

| Fase | Actividades clave | Entregables |
|------|-------------------|-------------|
| **1. Definición** | Identificar necesidades, definir KPIs, stakeholders | Documento de requisitos |
| **2. Diseño** | Modelar datos, diseñar arquitectura, mockups | Modelo dimensional, wireframes |
| **3. Desarrollo** | Implementar ETL, crear visualizaciones | Dashboard funcional |
| **4. Despliegue** | Testing, capacitación, lanzamiento | Sistema en producción |
| **5. Mantenimiento** | Monitorizar, optimizar, evolucionar | Mejoras iterativas |  

<br>  

> **Flujo:** Definición → Diseño → Desarrollo → Despliegue → Mantenimiento (ciclo continuo)
